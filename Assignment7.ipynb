```python

!pip install pdfplumber
import pdfplumber
import pandas as pd
from pdfplumber.utils import extract_text, get_bbox_overlap, obj_to_bbox
import tabulate

```

```python
def extract_text_from_pdf(path_to_pdf):
  all_text = []
  with pdfplumber.open(path_to_pdf) as pdf:
    for page_number, page in enumerate(pdf.pages, start=1):
      page_text = page.extract_text()
      if page_text:
                all_text.append(page_text)
      else:
          print(f"Warning: No text found on page {page_number}.")
  return "\n".join(all_text)
```

```python
pdf_path = r"/content/testdoc.pdf"
extracted_text = extract_text_from_pdf(pdf_path)
print(extracted_text)
```

```python
from nltk import word_tokenize, sent_tokenize
print(word_tokenize(extracted_text))
words = word_tokenize(extracted_text)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]
print(filtered_words)
```

```python
from nltk.stem.snowball import SnowballStemmer
porter = SnowballStemmer(language='english')
stemmed_words = [porter.stem(word) for word in filtered_words]
print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmed_words = [lemmatizer.lemmatize(word, 'v') for word in filtered_words]
print(lemmed_words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(words)
dense_matrix = tfidf_matrix.todense()
dense_array = np.array(dense_matrix)
words = tfidf_vectorizer.get_feature_names_out()
print("TF-IDF Matrix:")
print(dense_array)

print("\nFeature Names (Words):")
print(words)
```

```python
for i, doc in enumerate(words):
    print(f"\nDocument {i+1}: {doc}")
    word_scores = [(word, dense_array[i][j]) for j, word in enumerate(words)]


    word_scores.sort(key=lambda x: x[1], reverse=True)

    for word, score in word_scores:
        print(f"{word}: {score:.4f}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()


result = tfidf.fit_transform(words)


print('\nidf values:')

idf_values = list(zip(tfidf.get_feature_names_out(), tfidf.idf_))


idf_values.sort(key=lambda x: x[1], reverse=True)


for word, idf in idf_values:
    print(word, ':', idf)




print('\nWord indexes:')
print(tfidf.vocabulary_)


print('\ntf-idf value:')
print(result)

print('\ntf-idf values in matrix form:')
print(result.toarray())
```
